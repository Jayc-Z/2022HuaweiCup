```python
import numpy as np
from sklearn.decomposition import PCA
X1 =data5[[ 'SOC土壤有机碳', 'SIC土壤无机碳', 'STC土壤全碳', '全氮N', '土壤C/N比']].values
pca = PCA(n_components=1)   #降到1维
pca.fit(X1)                  #训练
newX1=pca.fit_transform(X1)   #降维后的数据
# PCA(copy=True, n_components=2, whiten=False)
print(pca.explained_variance_ratio_)  #输出贡献率
```

```python
X2 =data5[[ '10cm湿度(kg/m2)', '40cm湿度(kg/m2)', '100cm湿度(kg/m2)', '200cm湿度(kg/m2)']].values
pca = PCA(n_components=1)   #降到2维
pca.fit(X2)                  #训练
newX2=pca.fit_transform(X2)   #降维后的数据
# PCA(copy=True, n_components=2, whiten=False)
print(pca.explained_variance_ratio_)  #输出贡献率
```

```python
data5['有机物含量']=newX1
data5['土壤湿度']=newX2

data5['土壤板结化指数']=1/2*data5['有机物含量']+1/2*data5['土壤湿度']
i='土壤板结化指数'
data5[i]=(data5[i]-data5[i].min())/(data5[i].max()-data5[i].min())
```

```python
data5['jiehe']=1/2*data5['土壤板结化指数']+1/2*data5['沙漠化程度指数']
```

```python
#对照（NG， 0羊/天/公顷 ）、轻度放牧强度（LGI， 2羊/天/公顷 ）、中度放牧强度（MGI，4羊/天/公顷 ）和重度放牧强度（HGI，8羊/天/公顷 ）
data5['放羊数']=data5['放牧强度（intensity）'].map(
{
    'NG':0,
    'LGI':2, 
    'MGI':4, 
    'HGI':8
})
```

```python
import xgboost as xgb
import lightgbm as lgb
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
import numpy as np
from sklearn.model_selection import train_test_split
def mean_absolute_percentage_error(y_true, y_pred): 

    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100
```

```python
data5.columns
```

```python

X=data5[[ '降水量(mm)', 'jiehe']]

y=data5['放羊数']
# 拆分训练集和测试集
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=12)
    
#线性回归
model = LinearRegression()

model.fit(x_train, y_train)
print('线性回归')
print(mean_absolute_percentage_error(model.predict(x_test),y_test))


# Randomforest回归

model = RandomForestRegressor()

model.fit(x_train, y_train)
print('Randomforest回归')
print(mean_absolute_percentage_error(model.predict(x_test),y_test))

# lgbm分类
model = lgb.LGBMRegressor()
model.fit(x_train, y_train)
print('lgbm回归')
print(mean_absolute_percentage_error(model.predict(x_test),y_test))


# XGboost分类
model = xgb.XGBRFRegressor()
model.fit(x_train, y_train)
print('XGboost回归')
print(mean_absolute_percentage_error(model.predict(x_test),y_test))
```
