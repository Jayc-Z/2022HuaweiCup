```python
data['气象']=data[ '平均气温(℃)']+data[ '降水量(mm)']+data[ '平均风速(knots)']
data['地表']=data[ '10cm湿度(kg/m2)']+data[ '200cm湿度(kg/m2)']+data[ '植被指数(NDVI)']
data['沙漠化程度指数']=data['气象']+data['地表']
i='沙漠化程度指数'
data[i]=(data[i]-data[i].min())/(data[i].max()-data[i].min())
```

```python
data2=pd.read_excel('附件14不同放牧强度土壤碳氮监测数据.xlsx')
data2.head()
```

```python
data3=pd.read_excel('附件3、土壤湿度2022—2012年.xls')
data3
```

```python
data4=pd.merge(data2,data3,right_on=['年份'],left_on=['year'])

data4.reset_index(inplace=True,drop=True)
```

```python
data4=data4[['月份', '年份', '放牧小区（plot）', '放牧强度（intensity）', 'SOC土壤有机碳', 'SIC土壤无机碳',
       'STC土壤全碳', '全氮N', '土壤C/N比', 
       '10cm湿度(kg/m2)', '40cm湿度(kg/m2)', '100cm湿度(kg/m2)', '200cm湿度(kg/m2)']]
```

```python
data5=pd.merge(data4,data[['沙漠化程度指数','年份']],on=['年份'])
```

```python
data5
```

```python
data5.columns
```

```python
import numpy as np
from sklearn.decomposition import PCA
X1 =data5[[ 'SOC土壤有机碳', 'SIC土壤无机碳', 'STC土壤全碳', '全氮N', '土壤C/N比']].values
pca = PCA(n_components=1)   #降到1维
pca.fit(X1)                  #训练
newX1=pca.fit_transform(X1)   #降维后的数据
# PCA(copy=True, n_components=2, whiten=False)
print(pca.explained_variance_ratio_)  #输出贡献率
```

```python
X2 =data5[[ '10cm湿度(kg/m2)', '40cm湿度(kg/m2)', '100cm湿度(kg/m2)', '200cm湿度(kg/m2)']].values
pca = PCA(n_components=1)   #降到2维
pca.fit(X2)                  #训练
newX2=pca.fit_transform(X2)   #降维后的数据
# PCA(copy=True, n_components=2, whiten=False)
print(pca.explained_variance_ratio_)  #输出贡献率
```

```python
data5['有机物含量']=newX1
data5['土壤湿度']=newX2

data5['土壤板结化指数']=1/2*data5['有机物含量']+1/2*data5['土壤湿度']
i='土壤板结化指数'
data5[i]=(data5[i]-data5[i].min())/(data5[i].max()-data5[i].min())
```

```python
data5['jiehe']=1/2*data5['土壤板结化指数']+1/2*data5['沙漠化程度指数']
```
